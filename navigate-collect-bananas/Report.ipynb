## Implementation
The model and the agent is implemented in model.py and dqn_agent.py file respectively. The Navigation.ipynb notebook has a function dqn() to train the agent with default values for,
 - number of episodes - n_episodes
 - timesteps per episoddes - max_t
 - epsilon start value - eps_start
 - epsilon end value - eps_end
 - epsilon decay rate - eps_decay
 - minimum reward to consider the agent solved the environment - min_avg_reward
 
The agent is trained for 2000 episodes with 1000 timesteps per episode. Epsilon start and end values are 1.0 and 0.01 respectively and epsilon decay rate is 0.995. The minimum reward that agent should achieve to solve the environment is 13.

## Algorithm

Deep Q-Network used to train the agent using experienced replay and fixed Q-targets. The network is optimized using adam optimizer and mean squared error loss. Epsilon-greedy function is used to select the action with epsilon decay rate of 0.995. The given problem is an episodic task. For each episode, the agent is trained with maximum timesteps of 1000.

### Hyperparameters
The following hyperparameters used to train the agent.
- Replay buffer size, BUFFER_SIZE = 1e5
- Minibatch size, BATCH_SIZE = 64
- Discount factor, GAMMA = 0.99
- Target parameters soft update, TAU = 1e-3
- Learning rate, LR = 5e-4
- Target network update frequency, UPDATE_EVERY = 4

### Model architecture
Three fully connected layers neural network is used to map the state to action values. Rectified Linear unit (ReLU) activation function applied the layers except the output layer. The first two fully connected layers have output units of 64 neurons. The output layer has neurons of size equal to action size i.e 4

## Results

The agent solved the environment at episode 458 with average reward of 13.01 over last 100 episodes. The agent solved with best average reward of 16.58 at episode 1203.
